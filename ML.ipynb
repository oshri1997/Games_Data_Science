```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, f1_score
```

```python
df=pd.read_csv("clean_df_no_rounded.csv")
df
```

```python
df['Platforms amount'] = df.groupby('Platforms')['Platforms'].transform('count')
df['Publisher movie amount'] = df.groupby('Publisher')['Publisher'].transform('count')
df['Genre Amount'] = df.groupby('Genre')['Genre'].transform('count')
df
```

```python
X = df.drop(['Platforms','Publisher','Genre','Title','Release_date','Year','Total_sales'],axis=1)
y =  df['Total_sales']
```

```python
X.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
model = sklearn.linear_model.LinearRegression().fit(X_train, y_train)
```

```python
newDf = pd.DataFrame({'Total_Sales actual': y_test,'Total_Sales predicted':model.predict(X_test)})
```

```python
newDf.to_csv("Predicted.csv")
print(r2_score(y_test, newDf['Total_Sales predicted']))



```
