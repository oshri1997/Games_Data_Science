```python
import numpy as np
import pandas as pd
import matplotlib as plot
```

```python
def clean_dataframe(df):
    df = df.dropna(axis=0, subset=['Rating'])
    df=df.fillna(value=0)
    df = df.drop(df[df.Total_Sales==0].index)
    

    return df
```

```python
df=pd.read_csv("Games_Data_Test.csv",encoding='latin-1')
df
```

```python
cdf1=clean_dataframe(df)
cdf1.to_csv("clean_df_no_rounded.csv",index=False)
cdf1
```

```python
df['Total_sales'] = df['Total_sales'].apply(np.ceil)
df['Japan_Sales'] = df['Japan_Sales'].apply(np.ceil)
df['USA_Sales'] = df['USA_Sales'].apply(np.ceil)
df['EUR_Sales'] = df['EUR_Sales'].apply(np.ceil)
df
```

```python
cdf=clean_dataframe(df)
cdf.to_csv("clean_df.csv",index=False)
cdf
```
