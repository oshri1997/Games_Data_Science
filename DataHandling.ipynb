```python
import numpy as np
import pandas as pd
```

```python
def clean_dataframe(df):
    cleanDf=df.drop(df[df.Total_Sales<0.5].index)
    cleanDf["Total_Sales"]=round(df["Total_Sales"],1)
    cleanDf=cleanDf.dropna()
    return cleanDf
```

```python
df=pd.read_csv("Games_Data.csv")
df
```

```python
cdf1=clean_dataframe(df)
cdf1.to_csv("clean_df.csv",index=False)
cdf1
```
