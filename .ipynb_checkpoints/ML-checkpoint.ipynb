```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, f1_score
```

```python
df=pd.read_csv("test.csv")
df.head()
```

```python
df['Platform numer'] = df.groupby('Platforms')['Platforms'].transform('count')
df['Publisher movie numer'] = df.groupby('Publisher')['Publisher'].transform('count')
df['Genre numer'] = df.groupby('Genre')['Genre'].transform('count')
#def data_encode(x_data):
#   for i in x_data.columns:
#      if(type(x_data[i][0])==str):
#          x_data[i]=x_data[i].factorize()[0]
        
#return x_data    
    
#X= df.drop(['Total_sales','Release_date','Year'],axis=1)
#y =  df['Total_sales']
#X = data_encode(X)

#X
```

```python
X = df.drop(['Platforms','Publisher','Genre','Title','Year','Release_Date','Rating','EUR_Sales','Japan_Sales','USA_Sales'],axis=1)
y =  df['Rating']

X
```

```python
X.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
model = sklearn.linear_model.LinearRegression().fit(X_train, y_train)
```

```python
newDf = pd.DataFrame({'Rating actual': y_test,'Rating predicted':model.predict(X_test)})
```

```python
newDf.to_csv("Predicted.csv")

print(r2_score(y_test, newDf['Rating predicted']))



```
